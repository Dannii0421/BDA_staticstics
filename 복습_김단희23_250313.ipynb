```python
## 내가 필요한 수치형 데이터 타입만 추출하자
## select_types(include='number')
df_num = tt.select_dtypes(include='number')
```

```python
print('요약통계',df_num.describe())
print('평균',df_num.mean())
print('최빈값',df_num.mode())
print('사분위',df_num.quantile([0.25,0.5,0.75]))
print('사분위',df_num.quantile([0.3,0.5,0.9]))
print('상관계수',df_num.corr())
print('공분산',df_num.cov())
print('왜도',df_num.skew())
print('표준오차',df_num.sem())
print('유니크한 값',df_num.nunique())
print('인덱스기준 max',df_num.idxmax())
```

데이터 분석에 필요한 기초 문법들

-loc[],iloc[] 로케이션, 인덱스 위치
  * loc - location(인덱스) - 처음 생성된 데이터프레임의 인덱스
  * iloc - Integer location(넘버링된 인덱스) - 현재 기준의 데이터프레임의 순서, 데이터프레임을 재구조화 하면 그 기준으로 넘버링
  * loc[인덱스]
  * iloc[인덱스]
  * 안 되는 조건
      - loc[행,열], iloc[행,열]
      - loc[:,:], iloc[;,;]
      - 명시적으로 컬럼을 적어야 한다. loc
      - 숫자로만 컬럼을 적어야 한다. iloc
- query()
- df[df[불린]]
- sort_values()
- 파생변수만들기 - 판다스 문법,assign,lambda
- groupby
- merge,concat
- str
- 기타 stats 패키지 함수(상황에 따라)

- query 문법
   - query(조건)
   - query('변수 연산자 조건')
      - 주의할 점은 문자열은 "조건 연산자 '문자열'"

```python
## pclass 3번과 pclass 1번의 고객들의 남성과 여성의 각각 fare 평균을 나눠서 구하여라
pc3_male=tt[['pclass','fare','sex']].query("pclass==3 & sex == 'male'")['fare'].mean()
#tt[['pclass','fare','sex']].query("pclass==3 & sex == 'female'")
```

```python
pc3_male
```

```python
tt[['pclass','fare','sex']].query("pclass==1 & sex == 'male'")
tt[['pclass','fare','sex']].query("pclass==1 & sex=='female'")
```

df[df 비교연산자 불린]
 - 안에 들어가는 불린값이 참인 형태의 조건의 데이터만 추출이 된다
다중조건으로 사용하는 경우
 - df[(df조건1)&(df조건2)]

```python
tt
```

```python
## 다중 조건이 들어갈 수 있다.
## 20대 이상인데, pclass==3 인 경우는?
## & or
#tt[(tt['age']>20) & (tt['pclass']==3)][['fare','sex','pclass']].loc[0:100]['fare'].mean()

tt[(tt['age']>20) & (tt['pclass']==3)][['fare','sex','pclass']].loc[0:100]['fare']
```

```python
tt[(tt['age']>20) & (tt['pclass']==3)][['fare','sex','pclass']]
```
