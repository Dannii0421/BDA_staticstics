생존율을 바라보는 다양한 문법들 

클래스별로 생존에 따른 여러 테이블을 만들 수 있다.

- 교차테이블 crosstab()
  - 두 개 이상의 범주형 변수간의 빈도수를 측정한다.
  - 엑셀 피벗과 좀 유사한 내용
  - 각 범주에 따른 조합에 속하는 데이터 개수를 계산
- pd.crosstab(컬럼1, 컬럼2) 기준컬럼들이 들어간다.
  - pd.crosstab(컬럼1, 컬럼2, normalize= 'all') // 모든 값의 합이 1이되도록 비율을 계산한다.
  - pd.crosstab(컬럼1, 컬럼2, normalize= 'index') // 각 행의 합이 1이 되도록 정규화한다.

```python
pd.crosstab(tt['pclass'],tt['sex'],normalize='all')
```

```python
pd.crosstab(tt['pclass'],tt['sex'],normalize= 'index')
```

```python
pd.crosstab(tt['sex'],tt['survived'])
```

cut, qcut

- 구간화 binning 작업
  - 연속형 데이터를 -> 고정된 간격 (bins) 나눈다
  - 수치형 데이터를 범주형 데이터로 구간화 변환
- cut
  - 구간의 크기가 일정하게, 값의 분포를 고려하지 않는다. 구간 개수는 직접 우리가 설정 가능하다.
  - pd.cut(데이터프레임 컬럼, bins=지정)
- qcut
  - 가능한 데이터의 분포의 따라 자동으로 분위수 기준을 나누는 게 큰 차이
  - 데이터의 개수가 비슷한 그룹으로 나누고 싶을 때
  - 등간격이 아니라 분포를 고려해서 구간을 자동 설정
  - 구간에 속하는 데이터 개수가 동일하다.
  - pd.qcut(데이터프레임컬럼, q=지정)

```python
tt['cut']=pd.cut(tt['age'],bins=4)
```

```python
tt['qcut']=pd.qcut(tt['age'],q=4)
```

타이타닉 데이터를 확인했을 때
- 등급이 존재하고, 생존에 대한 생존자가 확인 가능하다.
- 타이타닉 데이터의 생존자와 비생존자의 등급별 생존율의 차이가 유의미한가? 유의미하지 않는가?

```python
tt.groupby('pclass')['survived'].sum()
```

- pclass 클래스와 생존에 따른 분석
- 범주형 형태로 카이제곱 통계량으로 분석할 수 있다.

```python
crss_table=pd.crosstab(tt['pclass'],tt['survived'])
```

```python
crss_table
```

```python
## sicpy에서 제공하는 chi2_contingency 함수로 쉽게 추출이 가능하다.
from scipy import stats

chi2, p, dof, expected = stats.chi2_contingency(crss_table)
```

```python
## p-value
if p<0.05:
    print('pclass와 생존율의 차이는 통계적으로 유의미하다')
else:
    print('pclass와 생존율의 차이는 통계적으로 유의미하지 않다.')
```

카이제곱의 간단한 정리

- 두 개 이상의 범주형 변수를 관계를 가지고 통계적인 방법으로 분석
- 카이제곱의 검정의 주요 용도
  - 독립성 : 두 범주형 변수간의 연관성 여부를 확인하는 것 등급(1,2,3)따라 생존여부(0,1) 통계적으로 유의미한 차이가 있는지? 분석
  - 적합성 : 관찰된 데이터가 기대되는 분포와 일치하는지 여부도 평가
    - 주사위 던지기
    - 1/6 기대빈도 -> 실제 주사위를 던지면서 나오는 관측된 빈도 일치하는지 이런 것을 보면서 분석
  - 카이제곱 검정 가설
    - 귀무가설(H0):두 변수는 독립적이다.(즉 연관성이 없다.)
    - 대립가설(H1):두 변수는 독립적이지 않다.(즉 연관성이 있다.)
  - 위에 추출된 값들을 확인하면
    - 관찰빈도
    - 기대빈도
    - 통계량
    - P-value
    - 결론 : p-value(0.05)보다 작으면 두 변수 간에 연관성이 있다는 결론
